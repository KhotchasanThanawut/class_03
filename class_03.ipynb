## 1) Function
---
A function is a reusable block of code that
> Reduce duplication, improve readability, and isolate logic.
- Accepts input (parameters)
- Executes logic
- Optionally returns output

### NOTE
- Functions do not share scope by default. 
- Inside the function runs only when called.
- Function without return = return `None`.
---

### Normal Code

```python
a = 1
b = 2
c = a + b

d = 3
e = 4ิ
f = d + e

print(c)
print(f)
```

### Code with Function

#### No `Return` 

```python
def menu():
    print("     Menu")
    print("+============+")
    print("1) Balancee")
    print("2) Food menu")
    print("3) Quit")
    print("+============+")

print("\n Function return:", menu())
```

#### Have `return`

```python
def add(a, b):
    return a + b

result1 = add(1, 2)
result2 = add(3, 4)

print(result1)
print(result2)
```

```python
def divide(a, b):
    if b == 0:
        return None
    return a / b

result = divide(1,3)
print(round(result, 3))
```

```python
def check_sign(n):
    if n > 0:
        return "positive"
    elif n == 0:
        return "zero"
    return "negative"

print(check_sign(1))
print(check_sign(-1))
print(check_sign(0))
print(check_sign(10))
```

```python
def greeting():
    print("Hello, my name is .....")
    print("I'm good, how's it going?")
    print("Goodbye mate.")

# greeting()
# print(greeting())
```

### Advance Usage

```python
def get_stats(nums):
    return min(nums), max(nums), sum(nums)

low, high, total = get_stats([1, 2, 3])

print(low, high, total)
```

## 2) Variables Scope
---
Scope defines where a variable is accessible.
- Local
- Global
- NonLocal

### Local vs Global

```python
x = "Global"

def test():
    x = "Local"
    print(x)

test()
print(x)
```

### Global Key

```python
x = "unchange"
# print(x)

def change():
    global x
    x = "new"

change()
print(x)
```

```python
# if no global???
x = 1

def change():
    x += 1

change() # Error cannot access local variable.
```

### Nonlocal

```python
def outer():
    x = 5
    def inner():
        nonlocal x
        x += 1
    inner()
    return x

print(outer())
```

## 3) if __name__ == "__main__"
---
Controls whether code runs when
- File is executed directly
- File is imported as a module

### NOTE
- Prevents unwanted execution during imports
- Essential for modular programs
- Mandatory in real projects

```python
# print("Hello world")

def main():
    print("Program running")

if __name__ == "__main__":
    main()
```

```python
def test():
    print("Testing logic 01")

def test2():
    print("Testing logic 02")

def what():
    print("Combine both \n")
    test()
    test2()

if __name__ == "__main__":
    what()
```
